```python
import matplotlib.pyplot as plt

def henon_map(a, b, x0, y0, n): #defining henon map function
    x, y = [x0], [y0]
    for _ in range(n):
        x_new = 1 - a * x[-1]**2 + y[-1]
        y_new = b * x[-1]
        x.append(x_new)
        y.append(y_new)
    return x, y

#parameters for the Hénon map
a = 1.4
b = 0.3
iterations = 500

#initial conditions as in time series
initial_conditions_1 = (0, 0)
initial_conditions_2 = (0.000001, 0)

#generate paths
x1, y1 = henon_map(a, b, initial_conditions_1[0], initial_conditions_1[1], iterations)
x2, y2 = henon_map(a, b, initial_conditions_2[0], initial_conditions_2[1], iterations)

#plotting
plt.figure(figsize=(10, 6))
plt.scatter(x1, y1, s=10, label='Initial condition (0,0)', color='red')
plt.scatter(x2, y2, s=10, label='Initial condition (0.000001,0)')
plt.title('Hénon Map with a=1.4, b=0.3')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
